```python
import numpy as np
import matplotlib.pyplot as plt
```

### Равномерное распределение 

```python
from scipy.stats import uniform

seg = np.linspace(-5, 5, 200)
plt.plot(seg, uniform.pdf(seg), color='red')
plt.title('Равномерное распределение \nПлотность случайной величины')
plt.grid()
```

```python
plt.plot(seg, uniform.cdf(seg), color='red')
plt.grid()
plt.title('Равномерное распределение \nФункция распределения случайной величины')
```

# Нормальное распределени

```python
from scipy.stats import norm

seg = np.linspace(-5, 5, 200)
plt.plot(seg, norm.pdf(seg), color='blue')
plt.grid()
plt.title('Нормальное распределение \nПлотность случайной величины')
```

```python
plt.plot(seg, norm.cdf(seg), color='blue')
plt.grid()
plt.title('Нормальное распределение \nФункция распределения случайной величины')
```

#  Распределение Пуассона

```python
from scipy.stats import poisson

seg = np.linspace(0, 30, 31)
plt.plot(seg, poisson(5).pmf(seg), color='blue')
plt.title('Распределение Пуассона \nПлотность случайной величины')
plt.grid()
```

```python
plt.step(seg, poisson(5).cdf(seg), color='blue')
plt.title('Распределение Пуассона \nФункция распределения случайной величины')
plt.grid()
```

# Экспоненциальное распределение

```python
from scipy.stats import expon

seg = np.linspace(-5, 5, 500)
plt.plot(seg, expon(2, 0.7).pdf(seg), color='blue')
plt.title('Распределение Пуассона \nПлотность случайной величины')
```

```python
plt.plot(seg, expon(2, 0.7).cdf(seg), color='blue')
plt.title('График функции распределения случайной величины')
```
